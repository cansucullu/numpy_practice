## Euclidean distance

**Formula**

![Euclidean distance formula](https://www.gstatic.com/education/formulas2/553212783/en/euclidean_distance.svg)

**Illustration**

![Illustration](https://upload.wikimedia.org/wikipedia/commons/5/55/Euclidean_distance_2d.svg)

````python
x = [3, 0]
y = [0, 4]
# We know the result should be 5
````

Calculate as a **Python list** and using built-in library **math**

````python
import math

def euclidean_distance_math(array1, array2):
    sum_of_sqaured_diffs = 0
    for i in range(len(array1)):
        sum_of_sqaured_diffs += (array1[i] - array2[i]) ** 2
    return math.sqrt(sum_of_sqaured_diffs)

euclidean_distance_math(x, y)
````

Calculate using **NumPy** array operations

````python
import numpy as np

xn = np.array(x)
yn = np.array(y)

def euclidean_distance_numpy(array1, array2):
    return np.sqrt(np.sum((array1 - array2) ** 2))

euclidean_distance_numpy(xn, yn)
````

Calculate using NumPy **linalg.norm()**

````python
def euclidean_distance_numpy_linalg(array1, array2):
  return np.linalg.norm(array1 - array2)

euclidean_distance_numpy_linalg(xn, yn)
````

Calculate using **scipy.spatial.distance.euclidean**

````python
from scipy.spatial.distance import euclidean

euclidean(xn, yn)
````

````python
def euclidean_distance_scipy(array1, array2):
    return euclidean(array1, array2)

euclidean_distance_scipy(xn, yn)
````

````python
from scipy.spatial.distance import euclidean, cdist
import numpy as np

point_a = [1, 2, 3]
point_b = [4, 5, 6]

# Using scipy.spatial.distance.euclidean
dist_scipy = euclidean(point_a, point_b)
print(f"Distance using scipy.spatial.distance.euclidean: {dist_scipy}")

# Using cdist for a distance matrix (requires input points to be array-like)
points_set1 = np.array([[1, 2], [3, 4]])
points_set2 = np.array([[5, 6], [7, 8]])

distance_matrix = cdist(points_set1, points_set2, 'euclidean')
print(f"Distance matrix using cdist:\n{distance_matrix}")
````

## Manhattan Distance

**Illustration**

![Manhattan Distance Illustration](https://www.maartengrootendorst.com/images/posts/2021-01-02-distances/manhattan.svg)

**Formula**

![Manhattan Distance Formula](https://www.maartengrootendorst.com/images/posts/2021-01-02-distances/manhattan_distance.png)

Using NumPy

````python
def manhattan_distance_numpy(array1, array2):
    return np.sum(abs(array1 - array2))

manhattan_distance_numpy(xn, yn)
````

Using Scipy

````python
from scipy.spatial.distance import cityblock

def manhattan_distance_scipy(array1, array2):
    return cityblock(array1, array2)

manhattan_distance_scipy(xn, yn)
````

## Chebyshev Distance

![](https://www.maartengrootendorst.com/images/posts/2021-01-02-distances/chebyshev.svg)

![](https://www.maartengrootendorst.com/images/posts/2021-01-02-distances/chebyshev_distance.png)

````python
def chebyshev_distance_numpy(array1, array2):
    return np.max(abs(array1 - array2))

chebyshev_distance_numpy(xn, yn)
````
