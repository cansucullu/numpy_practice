```python
import matplotlib.pyplot as plt
import numpy as np
```

## Histogram Example

```python
np.random.seed(1234)
x = np.random.normal(170, 10, 250)

plt.hist(x, bins=30)
plt.show()
```

```python
# From https://www.geeksforgeeks.org/python/matplotlib-pyplot-hist-in-python/

x = np.random.randn(10000, 3)
colors = ['green', 'blue', 'lime']

plt.hist(x, bins=20, density=True, histtype='bar', color=colors, label=colors)

plt.legend(fontsize=10)
plt.title('matplotlib.pyplot.hist() Example', fontweight='bold')

plt.show()
```

```python
# My custom version based on the tutorial above

x1 = np.random.normal(1, 1, 100000)
x2 = np.random.normal(3, 2, 100000)
x3 = np.random.normal(-5, 4, 100000)

x = np.stack((x1, x2, x3), axis=1)

colors = ['green', 'blue', 'lime']

plt.hist(x, bins=30, density=True, histtype='bar', color=colors, label=colors)

plt.legend(fontsize=10)
plt.title('matplotlib.pyplot.hist() Example', fontweight='bold')

plt.show()
```
